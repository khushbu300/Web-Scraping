```python
if not title:
    print("title not found")
title = title.text

summary = soup.find('header', 'article-header').find('h2')
if summary:
    summary = summary.text
else:
    summary = ""

content = soup.find('div', 'article-body').find_all('p')

dct =  {
    "title": title.strip(),
    "synopsis": summary.strip(),
    "content": [i.text.strip() for i in content]
}
```

```python
url = "https://www.bbc.com/news/world-europe-64142650"
body = requests.get(url)

soup = BeautifulSoup(body.content, 'html.parser')
soup = soup.find('article')
title = soup.find('h1')
if not title:
    print("no title")
title = title.text
synopsis = soup.find('b')
if synopsis:
    synopsis = synopsis.text
else:
    synopsis = ""
content = soup.find_all('p')

dct = {
    "title": title.strip(),
    "synopsis": synopsis.strip(),
    "content": [i.text.strip() for i in content]
}
```
